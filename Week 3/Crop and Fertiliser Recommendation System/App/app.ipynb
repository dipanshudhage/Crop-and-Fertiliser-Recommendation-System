```python

# Crop and Fertilizer Recommendation System (Google Colab Version)

# 📦 Install Required Packages
!pip install joblib

# 📁 Import Libraries
import pandas as pd
import joblib

# 📂 Upload model files
from google.colab import files

print("Upload: model_crop.pkl, label_encoder_crop.pkl, model_fertilizer.pkl, label_encoder_fertilizer.pkl")
uploaded = files.upload()

# 📌 Load Models and Label Encoders
crop_model = joblib.load("model_crop.pkl")
crop_le = joblib.load("label_encoder_crop.pkl")
fert_model = joblib.load("model_fertilizer.pkl")
fert_le = joblib.load("label_encoder_fertilizer.pkl")

# 🌾 Input Parameters (Manually or using input())
print("\nEnter Soil and Weather Parameters:")
N = int(input("Nitrogen (N): "))
P = int(input("Phosphorous (P): "))
K = int(input("Potassium (K): "))
ph = float(input("pH level: "))
temperature = float(input("Temperature (°C): "))
humidity = float(input("Humidity (%): "))
rainfall = float(input("Rainfall (mm): "))

# 🧠 Make Predictions
crop_input = pd.DataFrame([{
    'N': N,
    'P': P,
    'K': K,
    'temperature': temperature,
    'humidity': humidity,
    'ph': ph,
    'rainfall': rainfall
}])
crop_prediction = crop_model.predict(crop_input)
crop_name = crop_le.inverse_transform(crop_prediction)[0]

fert_input = pd.DataFrame([{
    'Nitrogen': N,
    'Potassium': K,
    'Phosphorous': P
}])
fert_prediction = fert_model.predict(fert_input)
fert_name = fert_le.inverse_transform(fert_prediction)[0]

# 📊 Display Results
print("\n✅ Recommended Crop:", crop_name)
print("✅ Recommended Fertilizer:", fert_name)
```
